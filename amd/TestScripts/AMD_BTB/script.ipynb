```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
repeat0 = 20
repeat1 = 1000
```

```python
final_df = pd.DataFrame()
for i in range(2,17):
    os.system("./run.sh {}".format(i))

    data = pd.read_csv("results.csv", comment='#', delim_whitespace=True)
    data["Clock"] = pd.to_numeric(data["Clock"],errors='coerce')
    data["CoreClockCnt"] = pd.to_numeric(data["CoreClockCnt"],errors='coerce')
    data["BrTaken"] = pd.to_numeric(data["BrTaken"],errors='coerce')
    data["BrMispred"] = pd.to_numeric(data["BrMispred"],errors='coerce')
    data["Clock"] = (data["Clock"]) / repeat1
    data["CoreClockCnt"] = data["CoreClockCnt"] / repeat1
    data["BrTaken"] = data["BrTaken"] / repeat1
    data["BrMispred"] = data["BrMispred"] / repeat1
    final_df[i] = data.groupby(data.index // repeat0).mean()["BrMispred"]
```

```python
final_df.plot()
```

```python
final_df.to_csv('final_df.csv')
```
